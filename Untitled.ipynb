```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime, timedelta 
from matplotlib import dates as mpl_dates
import datetime as dt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import plotly.offline as pyo
import folium
```

```python
df = pd.read_csv('data.csv')
df
```

```python
df.columns = df.columns.str.replace(r'\n','', regex=True)
df.columns
```

```python
df1 = df.groupby('DISTRICT')['TOTAL MEAT'].sum()
df1
```

```python
df2 = df1.drop(['NEPAL', 'FW.REGION', 'FW.TERAI', 'FW.HILLS', 'FW.MOUNTAIN', 'MW.REGION', 'MW.TERAI', 'MW.HILLS', 'MW.MOUNTAIN', 'W.REGION', 'W.TERAI', 'W.HILLS', 'W.MOUNTAIN', 'C.REGION', 'C.TERAI', 'C.HILLS', 'C.MOUNTAIN', 'E.REGION', 'E.TERAI', 'E.HILLS', 'E.MOUNTAIN'])
df2
```

```python
df2 = df2.reset_index()
df2
```

```python
District_geo = 'nepal.geojson'
District_geo
```

```python
data_to_plot = df2[['DISTRICT','TOTAL MEAT']]
data_to_plot.head()
```

```python
m = folium.Map(location=[28.3949, 84.1240], zoom_start=7.2)

folium.Choropleth(
    geo_data=District_geo,
    name='choropleth',
    data=data_to_plot,
    columns=['DISTRICT', 'TOTAL MEAT'],
    key_on='feature.properties.DISTRICT',
    fill_color='YlOrRd',
    fill_opacity=0.7,
    line_opacity=0.2,
    legend_name='TOTAL MEAT '
).add_to(m)


folium.LayerControl().add_to(m)

m
```
